Задача: написать DAG с последовательным выполнением двух простых задач, например, Bash команда echo "Hello" и команда date.

Цель: освоить структуру DAG, регистрацию задач, использование операторов (BashOperator, PythonOperator) и управление порядком задач через операторы >>.

Можно усложнить добавлением параллельных задач и зависимостей.

## Реализация ETL пайплайна

Задача: реализовать пример ETL (Extract-Transform-Load) процесса, где:

Задача 1 — извлечение данных (например, чтение CSV или API вызов)

Задача 2 — обработка и трансформация данных (чистка, агрегация, фильтрация)

Задача 3 — загрузка (запись в базу данных или файл)

Цель: понять, как связать задачи последовательно, обрабатывать ошибки, использовать PythonOperator, задавать параметры DAG (расписание, retries).
