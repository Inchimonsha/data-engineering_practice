## Построение системы Pub/Sub для обмена сообщениями

Задача: реализовать систему чатов или уведомлений в реальном времени через механизмы Publish/Subscribe Redis.

Задания: создать каналы, отправлять и получать сообщения через PUBLISH и SUBSCRIBE.

Подходит для построения систем с реальным временем обновления
