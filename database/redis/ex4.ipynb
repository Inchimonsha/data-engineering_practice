## Реализация системы авторизации на основе хранения сессий

Задача: хранить временные сессии пользователей с ограничением времени жизни (TTL).

Используются: строки с TTL (SET с параметром EX/ PX), управление временем жизни EXPIRE.

Позволяет быстро проверять авторизацию пользователей на сайте или в приложении
