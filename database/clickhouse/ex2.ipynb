## Анализ временных рядов и агрегаты

Постройте запросы на анализ временных рядов с использованием оконных функций, агрегатов (например, подсчёт скользящего среднего по времени).

Используйте функции toStartOfDay(), toStartOfHour() для группировки по времени.

Этот кейс полезен для работы с логами, метриками и любыми данными с отметкой времени.

Использование первичных индексов и сортировки данных

Практикуйтесь в выборе и написании правильного ORDER BY при создании таблиц типа MergeTree, чтобы понять, как работает первичный индекс ClickHouse и как он влияет на скорость выборок.

Экспериментируйте с различными вариантами порядка колонок в ORDER BY и сравнивайте время выполнения запросов.
